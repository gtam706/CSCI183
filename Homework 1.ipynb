```python
# Python Basics Q2
def findLargestProduct(nums):
    pair = [0, 0]
    max = 0
    prev = nums[0]
    
    for i in range(1, len(nums) - 1):
        if(prev * nums[i] > max):
            max = prev * nums[i]
            pair = [prev, nums[i]]
        prev = nums[i]
        
    return pair

nums = [8, 2, 4, 0, 9, 12, 3]
print(findLargestProduct(nums))
```

```python
#Python Basics Q3
def isPasswordValid(password): 
    if(len(password) < 6 or len(password) > 20):
        return "Should be between 6 to 20 characters long"
    
    validations = {'hasNum': False, 'hasUpper': False, 'hasLower': False, 'hasSpecial': False}
    
    # iterate through password
    for l in password:
        if(l >= '0' and l <= '9'):
            validations['hasNum'] = True
            continue
        if(l >= 'a' and l <= 'z'):
            validations['hasLower'] = True
            continue
        if(l >= 'A' and l <= 'Z'):
            validations['hasUpper'] = True
            continue
        if(l != ' '):
            validations['hasSpecial'] = True
            continue
    
    # check validations in dictionary
    if not validations['hasNum']:
        return "Should have at least one number"
    if not (validations['hasUpper'] and validations['hasLower']):
        return "Should have at least one uppercase and one lowercase character"
    if not validations['hasSpecial']:
        return "Should have at least one special symbol"
    
    return "Valid password"

print(isPasswordValid('pass'))
print(isPasswordValid("password"))
print(isPasswordValid("password1"))
print(isPasswordValid("Password1"))
print(isPasswordValid("Password1!"))
```

```python
#Python Basics Q4
def checkValidPassword(password):
    reasons = []
    stringDic=[]
    if len(password) < 9:
        reasons.append("Password is too short")
    if len(password) > 20:
        reasons.append("Password is too long")
    stor = ""
    count = 0
    current = ""
    for c in password:
        current += c
        if(len(current) == 2):
            if current in stringDic:
                reasons.append("Password has contains multiple same string pattern")
            else:
                stringDic.append(current)
            current = ""
        if(stor == c):
            count+=1
        else:
            count = 1
            stor = c
        if(count > 3):
            reasons.append("Password has 3 or more repeating characters in a row")
        if c == " ":
            reasons.append("Password has a newline or space")
    if len(reasons) > 0:
        print("Password is weak for following reasons: ")
        print(reasons)
    else:
        print("Password is strong")
checkValidPassword("aaaa")

```

```python
#Python for Data Science Q6
def n_len_words(fname, n):
    valid_words = set()
    with open(fname, "r") as file:
        for line in file.readlines():
            for word in line.split():
                if len(word) == n:
                    valid_words.add(word.lower())
    return valid_words

if __name__ == "__main__":
    print(n_len_words("./Downloads/Data Science.txt", int(input())))
    
```

```python
# Regular Expressions Q1
import re
def zchecker(word):
    match = re.match("[^zZ]\w*[^zZ]",word)
    if match is None:
        print("Not a valid stirng")
    else:
        print("Valid string")
zchecker("aza")
```
