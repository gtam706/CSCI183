```python
import pandas
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
import seaborn as sns
```

```python
pd = pandas.read_csv('vgsales.csv')
print('Column Names\tData Types')
print(pd.dtypes)
```

```python
pd.corr(method='pearson')
```

```python
genereFrame = pd.groupby(['Year','Genre'],as_index=False)['Global_Sales'].sum()
print('Column Names\tData Types')
print(genereFrame.dtypes)
print(genereFrame)
```

```python
sns.set(rc={'figure.figsize':(10,50)})
sns.barplot(data=genereFrame,x='Global_Sales',y='Year',hue='Genre',orient ='h',palette='bright')
```

```python
df = pd.groupby(['Platform'],as_index=False)['Global_Sales'].sum()
Gdf = df.sort_values('Global_Sales',ascending =False)
sns.set(rc={'figure.figsize':(10,10)})
sns.barplot(data=Gdf.head(),x='Platform',y='Global_Sales')
```
