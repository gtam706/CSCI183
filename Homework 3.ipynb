```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns
heartData = pd.read_csv('heart.csv')
```

```python
numericalFeatures = ['age','trestbps','thalach','oldpeak','chol']
ax = 0
fig, axes = plt.subplots(10, 1,figsize = (15,50))
for i in range(0,5):
    for j in range(i+1,5):
        plot = sns.scatterplot(ax=axes[ax],data=heartData,x=numericalFeatures[i],y=numericalFeatures[j],hue='target')
        ax = ax + 1
```

```python
X = heartData.drop(['sex','cp','fbs','restecg','exang','slope','thal','target','ca','oldpeak','age','trestbps'],axis = 1)
y = heartData['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.7, random_state = 0)
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
y_pred = log_reg.predict(X_test)
y_pred
b = log_reg.intercept_[0]
w1,w2 = log_reg.coef_.T
c = -b/w2
m = -w1/w2
xmin, xmax = 0, 600
ymin, ymax = 0, 250
xd = np.array([xmin, xmax])
yd = m*xd + c
plt.plot(xd, yd, 'k', lw=1, ls='--')

plt.xlim(xmin, xmax)
plt.ylim(ymin, ymax)
plt.ylabel('Maximum heart rate achieved')
plt.xlabel('serum cholestoral in mg/dl')
sns.scatterplot(data=heartData,x=heartData['chol'],y=heartData['thalach'],hue='target')
plt.show()

print(accuracy_score(y_pred, y_test))
print(confusion_matrix(y_pred, y_test))
```
